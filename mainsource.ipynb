```python
import bs4
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import requests
import seaborn as sns

#Past 5 seasons QB?

# Losing sucks. Nobody wants to do that shit. What do you do when your football team loses? Draft a Qb. How often do teams 
# with QBs losing most of their games stick with their QB? Does it work out for them? How often do they not? Does it work for them?



url_2015 = "https://www.pro-football-reference.com/years/2015/"
df_afc_2015 = pd.read_html(url_2015)[0]
df_nfc_2015 = pd.read_html(url_2015)[1]

df_afc_2015_ls = df_afc_2015[df_afc_2015.index % 5 != 0]
df_nfc_2015_ls = df_nfc_2015[df_nfc_2015.index % 5 != 0]
frames = [df_afc_2015_ls, df_nfc_2015_ls]
ls_2015 = pd.concat(frames, ignore_index=True)
ls_2015['W'] = ls_2015['W'].astype('int')
ls_2015['Year'] = 2015


url_2016 = "https://www.pro-football-reference.com/years/2016/"
df_afc_2016 = pd.read_html(url_2016)[0]
df_nfc_2016 = pd.read_html(url_2016)[1]

df_afc_2016_ls = df_afc_2016[df_afc_2016.index % 5 != 0]
df_nfc_2016_ls = df_nfc_2016[df_nfc_2016.index % 5 != 0]
frames = [df_afc_2016_ls, df_nfc_2016_ls]
ls_2016 = pd.concat(frames, ignore_index=True)
ls_2016['W'] = ls_2016['W'].astype('int')
ls_2016['Year'] = 2016


url_2017 = "https://www.pro-football-reference.com/years/2017/"
df_afc_2017 = pd.read_html(url_2017)[0]
df_nfc_2017 = pd.read_html(url_2017)[1]

df_afc_2017_ls = df_afc_2017[df_afc_2017.index % 5 != 0]
df_nfc_2017_ls = df_nfc_2017[df_nfc_2017.index % 5 != 0]
frames = [df_afc_2017_ls, df_nfc_2017_ls]
ls_2017 = pd.concat(frames, ignore_index=True)
ls_2017['W'] = ls_2017['W'].astype('int')
ls_2017['Year'] = 2017


url_2018 = "https://www.pro-football-reference.com/years/2018/"
df_afc_2018 = pd.read_html(url_2018)[0]
df_nfc_2018 = pd.read_html(url_2018)[1]

df_afc_2018_ls = df_afc_2018[df_afc_2018.index % 5 != 0]
df_nfc_2018_ls = df_nfc_2018[df_nfc_2018.index % 5 != 0]
frames = [df_afc_2018_ls, df_nfc_2018_ls]
ls_2018 = pd.concat(frames, ignore_index=True)
ls_2018['W'] = ls_2018['W'].astype('int')
ls_2018['Year'] = 2018


url_2019 = "https://www.pro-football-reference.com/years/2019/"
df_afc_2019 = pd.read_html(url_2019)[0]
df_nfc_2019 = pd.read_html(url_2019)[1]

df_afc_2019_ls = df_afc_2019[df_afc_2019.index % 5 != 0]
df_nfc_2019_ls = df_nfc_2019[df_nfc_2019.index % 5 != 0]
frames = [df_afc_2019_ls, df_nfc_2019_ls]
ls_2019 = pd.concat(frames, ignore_index=True)
ls_2019['W'] = ls_2019['W'].astype('int')
ls_2019['Year'] = 2019


years_list = [ls_2015, ls_2016, ls_2017, ls_2018, ls_2019]
years_frame_ls = pd.concat(years_list, ignore_index=True)

years_frame_ls
```

```python
#Creates a seaborn violin plot by taking the continent and added residuals from the data chart
ax = sns.violinplot(x = years_frame_ls.Year, y = years_frame_ls.W, data = years_frame_ls)

#Sets labels on the chart
ax.set_title("Wins by Years")
ax.set_ylabel("Wins")
ax.set_xlabel("Years")
```

```python
years_frame_ls['W-L%'] = years_frame_ls['W-L%'].astype('float')
years_frame_ls['Tm'] = [name.replace("+", "").replace("*", "") for name in years_frame_ls['Tm']]

teamnamedict = {'New England Patriots' : 'nwe', 'New York Jets' : 'nyj', 'Buffalo Bills' : 'buf', 'Miami Dolphins' : 'mia', 'Cincinnati Bengals' : 'cin',
    'Pittsburgh Steelers' : 'pit', 'Baltimore Ravens' : 'rav', 'Cleveland Browns' : 'cle', 'Houston Texans' : 'htx', 'Indianapolis Colts' : 'clt',
    'Jacksonville Jaguars' : 'jax', 'Tennessee Titans' : 'oti', 'Denver Broncos' : 'den', 'Kansas City Chiefs' : 'kan', 'Oakland Raiders' : 'rai',
    'San Diego Chargers' : 'sdg', 'Washington Redskins' : 'was', 'Philadelphia Eagles' : 'phi', 'New York Giants' : 'nyg', 'Dallas Cowboys' : 'dal',
    'Minnesota Vikings' : 'min', 'Green Bay Packers' : 'gnb', 'Detroit Lions' : 'det', 'Chicago Bears' : 'chi', 'Carolina Panthers' : 'car',
    'Atlanta Falcons' : 'atl', 'New Orleans Saints' : 'nor', 'Tampa Bay Buccaneers' : 'tam', 'Arizona Cardinals' : 'crd', 'Seattle Seahawks' : 'sea',
    'St. Louis Rams' : 'ram', 'San Francisco 49ers' : 'sfo', 'Los Angeles Rams' : 'ram'
}

low_years_frame_ls = years_frame_ls[years_frame_ls['W-L%'] <= .250]
low_years_frame_ls = low_years_frame_ls.reset_index(drop=True)
low_years_frame_ls
```

```python

i = 0
while i <= len(low_years_frame_ls.index) - 1:
    low_years_frame_ls.loc[i, 'Team_Name_and_Year'] = str(low_years_frame_ls["Tm"].iloc[i]) + " " + str(low_years_frame_ls["Year"].iloc[i])
    i = i + 1

low_years_frame_ls
```

```python
ax1 = sns.barplot(x = low_years_frame_ls.Team_Name_and_Year, y = low_years_frame_ls.W, data = low_years_frame_ls, palette=("hls"))
for item in ax1.get_xticklabels():
    item.set_rotation(90)

ax1.set_title("Win Values of Teams under .250 W/L from 2015 - 2019")
ax1.set_ylabel("Total Wins")
ax1.set_xlabel("Team Names and Years")
```

Wha happen QB?

```python
i = 0
while i <= len(low_years_frame_ls.index) - 1:
    url = "https://www.pro-football-reference.com/teams/" +  teamnamedict[str(low_years_frame_ls["Tm"].iloc[i])] + "/" + str(low_years_frame_ls["Year"].iloc[i]) + "_roster.htm"
    currtab = pd.read_html(url)[0]
    low_years_frame_ls.loc[i ,'CurrentQB'] = currtab['Player'][1]
    i = i + 1

low_years_frame_ls
```

```python
i = 0
while i <= len(low_years_frame_ls.index) - 1:
    url = "https://www.pro-football-reference.com/teams/" +  teamnamedict[str(low_years_frame_ls["Tm"].iloc[i])] + "/" + str(low_years_frame_ls["Year"].iloc[i] + 1) + "_roster.htm"
    currtab = pd.read_html(url)[0]
    low_years_frame_ls.loc[i ,'NextYearQB'] = currtab['Player'][1]
    i = i + 1

low_years_frame_ls
```

```python

url_2020 = "https://www.pro-football-reference.com/years/2020/"
df_afc_2020 = pd.read_html(url_2020)[0]
df_nfc_2020 = pd.read_html(url_2020)[1]

df_afc_2020_ls = df_afc_2020[df_afc_2020.index % 5 != 0]
df_nfc_2020_ls = df_nfc_2020[df_nfc_2020.index % 5 != 0]
frames = [df_afc_2020_ls, df_nfc_2020_ls]
ls_2020 = pd.concat(frames, ignore_index=True)
ls_2020['W'] = ls_2020['W'].astype('int')
ls_2020['Year'] = 2020


updated_years_list = [ls_2015, ls_2016, ls_2017, ls_2018, ls_2019, ls_2020]
updated_years_frame_ls = pd.concat(updated_years_list, ignore_index=True)
updated_years_frame_ls['Tm'] = [name.replace("+", "").replace("*", "") for name in updated_years_frame_ls['Tm']]
updated_years_frame_ls.loc[176 ,'Tm'] = "Washington Redskins"


updated_years_frame_ls

```

```python
i = 0
while i <= len(low_years_frame_ls.index) - 1:
    low_years_frame_ls.loc[i ,'NextYearWins'] = updated_years_frame_ls.loc[(updated_years_frame_ls['Year'] == int(str(low_years_frame_ls["Year"].iloc[i] + 1))) & (updated_years_frame_ls['Tm'] == str(low_years_frame_ls["Tm"].iloc[i]))]['W'].item()
    i = i + 1

low_years_frame_ls
```

```python
ax1 = sns.countplot(x='NextYearWins', data=low_years_frame_ls)
for item in ax1.get_xticklabels():
    item.set_rotation(45)

ax1.set_title("Wincounts of Teams the Following Year")
ax1.set_ylabel("Number of Teams with Win Count")
ax1.set_xlabel("Number of Wins the Following Year")
```
